```python
def preprocess_input(filename, folder=False):
    # master list of notes
    notes = []
    
    # converting folders with multiple MIDI files
    if folder == True:
        for file in glob.glob('../input/classical-music-midi/'+filename+'/*.mid'):
            
            # read the MIDI file
            midi = converter.parse(file)            
            
            try: # file has instrument parts
                s2 = instrument.partitionByInstrument(midi)
                notes_to_parse = s2.parts[1].recurse() 
            except: # file has notes in a flat structure
                notes_to_parse = midi.flat.notes
            
#             print(notes_to_parse)
            for element in notes_to_parse:
                if isinstance(element, note.Note):
                    notes.append(str(element.pitch))
                elif isinstance(element, chord.Chord):
                    notes.append('.'.join(str(n) for n in element.normalOrder))
    else:
        midi = converter.parse(filename)
        try: # file has instrument parts
            s2 = instrument.partitionByInstrument(midi)
            notes_to_parse = s2.parts[1].recurse() 
        except: # file has notes in a flat structure
            notes_to_parse = midi.flat.notes
#         print(notes_to_parse)
        for element in notes_to_parse:
            if isinstance(element, note.Note):
                notes.append(str(element.pitch))
            elif isinstance(element, chord.Chord):
                notes.append('.'.join(str(n) for n in element.normalOrder))
        

    with open('./notes', 'wb') as filepath:
        pickle.dump(notes, filepath)
#     print(notes)
    return notes
         
```

```python
# a one hot encoding of the notes
def encode(notes):
    sortnotes = sorted(set(note for note in notes))
    dictnotes = dict((note, num) for (num, note) in enumerate(sortnotes))
    
    return dictnotes
```

```python
# in a RNN we treat sequences, so we will be reading a sequence of 100 notes in a MIDI file, and then predict the 101th note
# for this we need to create a dataset of a sequence and its next note, in the following function
def create_dataset(notes, n_notes, seqsize):
    
    listinput = []
    listoutput = []
    
    encnotes = encode(notes)
    # we will be looping till the last 101th sequence
    for i in range(0, len(notes) - seqsize, 1):
        inputs = notes[i:i+seqsize]
        output = notes[i+seqsize]
        
        listinput.append([encnotes[note] for note in inputs])
        listoutput.append(encnotes[output])
    
    
    # this isnt in the correct shape
    listinput = np.reshape(listinput, [len(listinput), seqsize, 1])
    print("New shape: "+ str(listinput.shape))
    # normalising the vector to make calculations more streamlined
    listinput = listinput / float(n_notes)

    
    # output as categorical data
    listoutput = np_utils.to_categorical(listoutput)
    
    return listinput, listoutput
        
```

```python
ip, op = create_dataset(notes, n_notes, 100)
```

# Model

```python
# creating model

def create_model(inputs, endsize):
    
    # using Keras
    model = Sequential()
    # input shape is the (seqsize, 1) per input
    model.add(layers.LSTM(
        256,
        input_shape=(inputs.shape[1], inputs.shape[2]),
        return_sequences=True,
        use_bias=False
    ))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(0.3))
#     model.add(layers.LSTM(512, return_sequences=True, use_bias=False))
#     model.add(layers.Dropout(0.3))
    model.add(layers.LSTM(256, return_sequences=True))
    model.add(layers.Dropout(0.3))
    model.add(layers.LSTM(256))
    model.add(layers.Dense(256))
    model.add(layers.Dropout(0.3))
    model.add(layers.Dense(endsize))
    model.add(layers.Activation('softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='rmsprop')

    return model
```

```python
model = create_model(ip, n_notes)
```

# Train

```python
# keeping checkpoints so I dont lose the weights
def train_network(filename):
    filepath = "./weight"+filename+".hdf5"

    checkpoint = ModelCheckpoint(
        filepath,
        monitor='loss',
        verbose=0,
        save_best_only=True,
        mode='min'
    )
    callbacks_list = [checkpoint]

    model.fit(ip, op, epochs=200, batch_size=120, callbacks=callbacks_list)
```

```python
train_network('bach_all')
```

# Generating Music

```python
#load the notes used to train the model
with open('./notes', 'rb') as filepath:
    notes = pickle.load(filepath)
```

```python
# a function to generate music
def generate(model, inputs, notes, n_notes, seqlength=100):
    
    decodenotes = dict((num, note) for (num, note) in enumerate(notes))
    
    # we begin with a random sequence from the inputs that we have
    # to get the index
    init = np.random.randint(0, len(inputs)-1)
    
    # the sequence 
    seq = inputs[init]
    print(type(seq))
#     seq = [note for note in list for list in seq]
    listoutput = []
    seq = np.concatenate(seq).ravel().tolist()
    for i in range(seqlength):
        # reshaping and normalising the input
        inp = np.reshape(seq, (1, len(seq), 1))
        inp = inp / float(n_notes)
        # getting the prediction
        pred = model.predict(inp, verbose=0)
        
        # the returned is an encoding of a note
        n = np.argmax(pred)
        pred = decodenotes[n]
        
        # adding the note to the sequence
        listoutput.append(pred)
        
        # now we slide the window to the right by a step for the next prediction
        seq = seq[1:len(seq)]
        seq.append(n)

    return listoutput
```

```python
new_gen = generate(model, ip, notes, n_notes, seqlength=400)
```

```python
# Creating a MIDI file from prediction
def create_MIDI(gen, name=""):
    
    # the offset is the time difference between notes, we assume its 0.5 here
    offset = 0
    music = []
    
    for seq in gen:
        # chords are seperated by .
        if ('.' in seq) or seq.isdigit():
            chordnotes = seq.split('.')
            n = []
            
            for cur_note in chordnotes:
                new_n = note.Note(int(cur_note))
                new_n.storedInstrument = instrument.Piano() # single piano instrument only
                n.append(new_n)
            
            new_c = chord.Chord(notes)
            new_c.offset = offset
            music.append(new_c)
        
        else:            
            new_n = note.Note(seq)
            new_n.storedInstrument = instrument.Piano() # single piano instrument only
            new_n.offset = offset
            music.append(new_n)
        
        offset += 0.5
    
    print(notes)
    # producing a MIDI stream
    midi = stream.Stream(music)
    
    midi.write('midi', fp='new_music_'+name+'.mid')
```
